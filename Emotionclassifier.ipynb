EMOTION CLASSIFIER

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

Read File

```python
file = open('emotionClassifier.txt','r')
data = file.readlines()
```

```python
data
```

```python
c1 = []
c2 = []
```

```python
for i in data:
  s = i.split(';')
  c1.append(s[0])
  c2.append(s[1])
```

```python
c2
```

Dataset

```python
w = {'Statement' : c1 ,'Emotion' : c2}
```

```python
df = pd.DataFrame(w)
```

```python
df
```

```python
df['Emotion'] = df['Emotion'].str.replace('\n','')
```

```python
df
```

```python
df.info()
```

```python
df
```

Importing Nlp modules for text preprocessing

```python
import nltk
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("stopwords")

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
def cleantext(text):
  tokens = word_tokenize(text.lower())
  ftoken = [t for t in tokens if(t.isalpha())]
  stop = stopwords.words("english")
  ctoken = [t for t in ftoken if(t not in stop)]
  lemma = WordNetLemmatizer()
  ltoken = [lemma.lemmatize(t) for t in ctoken]
  return " ".join(ltoken)
```

```python
df["cleantext"]=df["Statement"].apply(cleantext)
```

```python
df
```

Data Separation 

```python
x = df.cleantext
y = df.Emotion
```

Creating new column of sentence count

```python
s = []
for i in df['cleantext']:
  s.append(len(word_tokenize(i)))
```

```python
df['Sentence'] = s
```

```python
df.head()
```

```python
max(df.Sentence)
```

```python
np.quantile(s, 0.95)
```

```python
max_len = np.quantile(s, 0.95)
```

Tokenizer Class

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
```

```python
tok = Tokenizer(char_level=False, split=" ")

tok.fit_on_texts(x)
tok.index_word
```

Length of vocabulary

```python
vocab_len = len(tok.index_word)
vocab_len
```

Sequence of text

```python
seq = tok.texts_to_sequences(x) 
seq
```

Padding

```python
seqmat = sequence.pad_sequences(seq, maxlen= int(max_len))
seqmat 
```

```python
vocab_len
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(seqmat,y,test_size=0.3,random_state=1)
```

Neccessary Classes for Rnn model buliding

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding,Dropout
```

Creating rnn model

```python
rnn = Sequential()
```

Adding Embedding layer

```python
rnn.add(Embedding(vocab_len+1,400, input_length=int(max_len), mask_zero=True))
```

RNN

```python
rnn.add(SimpleRNN(units=32, activation="tanh"))
rnn.add(Dropout(0.2))
```

Hidden layer

```python
rnn.add(Dense(units=32, activation="relu"))
rnn.add(Dropout(0.2))
```

```python
rnn.add(Dense(units=10, activation="softmax"))
```

```python
rnn.compile(optimizer="adam", loss="sparse_categorical_crossentropy",metrics=['accuracy'])
```

Fitting training data

```python
rnn.fit(xtrain, ytrain, batch_size=70, epochs=15)
```

Prediction

```python
ypred = rnn.predict(xtest)
```

```python
ypred
```

```python
yp = ypred.argmax(axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,yp))
```
